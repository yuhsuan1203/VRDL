```python
import keras
import matplotlib.pyplot as plt
from keras.layers import Dense, GlobalAveragePooling2D, Dropout, Activation, Flatten
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Model
from keras.optimizers import Adam, SGD
from keras.applications.vgg16 import VGG16
from keras.applications.resnet50 import ResNet50

print("Import Keras")
```

```python
### Specify Training Parameters
batchSize = 64               #-- Training Batch Size
num_classes = 13             #-- Number of classes
num_epochs = 25              #-- Number of epochs for training   
#learningRate= 0.001          #-- Learning rate for the network
#lr_weight_decay = 0.95       #-- Learning weight decay. Reduce the learn rate by 0.95 after epoch
```

```python
model = VGG16(weights='imagenet', include_top=False, input_shape=(256, 256, 3))

x = model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x) #dense layer 3
x = Dropout(0.45)(x)
x = Dense(128, activation='relu')(x) #dense layer 5
x = Dropout(0.45)(x)
preds = Dense(num_classes, activation='softmax')(x) #final layer with softmax activation

model2 = Model(inputs=model.input, outputs=preds)
for layer in model2.layers[:20]:
    layer.trainable=False
for layer in model2.layers[20:]:
    layer.trainable=True
    
model2.summary()
```

```python
model3 = VGG16(weights='imagenet', include_top=False, input_shape=(256, 256, 3))

x3 = model3.output
x3 = GlobalAveragePooling2D()(x3)
x3 = Dense(512, activation='relu')(x3) #dense layer 3
x3 = Dropout(0.42)(x3)
x3 = Dense(128, activation='relu')(x3) #dense layer 5
x3 = Dropout(0.42)(x3)
preds3 = Dense(num_classes, activation='softmax')(x3) #final layer with softmax activation

model4 = Model(inputs=model3.input, outputs=preds3)
for layer in model4.layers[:20]:
    layer.trainable=False
for layer in model4.layers[20:]:
    layer.trainable=True
    
model4.summary()
```

```python
model5 = VGG16(weights='imagenet', include_top=False, input_shape=(256, 256, 3))

x5 = model5.output
x5 = GlobalAveragePooling2D()(x5)
x5 = Dense(512, activation='relu')(x5) #dense layer 3
x5 = Dropout(0.39)(x5)
x5 = Dense(128, activation='relu')(x5) #dense layer 5
x5 = Dropout(0.39)(x5)
preds5 = Dense(num_classes, activation='softmax')(x5) #final layer with softmax activation

model6 = Model(inputs=model5.input, outputs=preds5)
for layer in model6.layers[:20]:
    layer.trainable=False
for layer in model6.layers[20:]:
    layer.trainable=True
    
model6.summary()
```

```python
from keras.applications.imagenet_utils import preprocess_input
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input, horizontal_flip=True, validation_split=0.12) #included in our dependencies

train_generator = train_datagen.flow_from_directory('/kaggle/input/cs-ioc5008-hw1/dataset/dataset/train/', # specify the path to the main data folder
                                                   target_size=(256, 256),
                                                   batch_size=batchSize,
                                                   class_mode='categorical',
                                                   shuffle=True,
                                                   subset='training')

train_datagen2 = ImageDataGenerator(preprocessing_function=preprocess_input, horizontal_flip=True, validation_split=0.1) #included in our dependencies

train_generator2 = train_datagen2.flow_from_directory('/kaggle/input/cs-ioc5008-hw1/dataset/dataset/train/', # specify the path to the main data folder
                                                     target_size=(256, 256),
                                                     batch_size=batchSize,
                                                     class_mode='categorical',
                                                     shuffle=True,
                                                     subset='training')

train_datagen3 = ImageDataGenerator(preprocessing_function=preprocess_input, horizontal_flip=True, validation_split=0.08) #included in our dependencies

train_generator3 = train_datagen3.flow_from_directory('/kaggle/input/cs-ioc5008-hw1/dataset/dataset/train/', # specify the path to the main data folder
                                                     target_size=(256, 256),
                                                     batch_size=batchSize,
                                                     class_mode='categorical',
                                                     shuffle=True,
                                                     subset='training')
```

```python
label_map = train_generator.class_indices
label_map = dict((v,k) for k,v in label_map.items())
for i in range(len(label_map)):
    print("class" + " {:>2}".format(str(i+1)) + ": " + label_map[i])
    
```

```python
model2.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
model4.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
model6.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
# Adam optimizer
# loss function will be categorical cross entropy
# evaluation metric will be accuracy

step_size_train = train_generator.n // train_generator.batch_size
step_size_train2 = train_generator2.n // train_generator2.batch_size
step_size_train3 = train_generator3.n // train_generator3.batch_size

model2.fit_generator(generator=train_generator, 
                     steps_per_epoch=step_size_train,  
                     epochs=num_epochs)

model4.fit_generator(generator=train_generator2, 
                     steps_per_epoch=step_size_train2, 
                     epochs=num_epochs)

model6.fit_generator(generator=train_generator3, 
                     steps_per_epoch=step_size_train3, 
                     epochs=num_epochs)
```
